```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
import numpy as np
```

```python
df = pd.read_csv(r"Medical Appointments No-Show/medical-appointments-no-show-en.csv")
```

```python
df
```

```python
df.isna().sum()
```

```python
print('Duplicate rows:',df.duplicated().sum())
```

```python
print("Numerical Summary:\n")
df.describe()
```

```python
print(" Categorical Summary:\n")
df.describe(include='object')
```

```python
if 'no_show' in df.columns:
    print("Value counts for target variable:")
    print(df['no_show'].value_counts())
    sns.countplot(data=df, x='no_show')
    plt.title("Distribution of No-Show Appointments")
    plt.show()
```

```python
numeric_cols = df.select_dtypes(include=[np.number]).columns

df[numeric_cols].hist(bins=20, figsize=(12, 10))
plt.suptitle("Numeric Features Distribution", fontsize=14)
plt.show()
```

```python
plt.figure(figsize = (10,9))
corr = df.corr(numeric_only=True)
sns.heatmap(corr, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title("Correlation Heatmap")
plt.show()
```

```python
cat_cols = df.select_dtypes(include=['object', 'category']).columns

for col in cat_cols:
    if col != 'no_show' and df[col].nunique() < 15:
        plt.figure(figsize=(6, 4))
        sns.countplot(data=df, x=col, hue='no_show')
        plt.title(f"{col} vs No-Show")
        plt.xticks(rotation=45)
        plt.tight_layout()
        plt.show()
```

```python
df_copy = df.copy()
```

```python
sns.scatterplot(data=df, x='age', y='average_temp_day', hue='no_show', alpha=0.6)
plt.title("Age vs Average Temperature — Colored by No-Show")
plt.tight_layout()
plt.show()
```

```python
sns.scatterplot(data=df, x='average_temp_day', y='average_rain_day', hue='no_show', alpha=0.6)
plt.title("Average Temperature vs Average Rain — Colored by No-Show")
plt.tight_layout()
plt.show()
```

```python
cols_for_pairplot = [
    'age', 'average_temp_day', 'average_rain_day', 
    'max_temp_day', 'max_rain_day', 'rain_intensity', 'no_show'
]

sns.pairplot(df[cols_for_pairplot], hue='no_show', diag_kind='kde')
plt.suptitle("Pairwise Relationships (Colored by No-Show)", y=1.02)
plt.show()
```
